```python
import json

# Merge the datasets on 'Name'
merged_data = pd.merge(michelin_data, google_data, on="Name", how="inner")
# Ensure NaN values are replaced with None for MongoDB JSON
merged_data = merged_data.where(pd.notna(merged_data), None)

# Convert DataFrame to structured JSON
structured_data = []

for _, row in merged_data.iterrows():
    structured_entry = {
        "Name": row["Name"],
        "michelin_info": {
            "Address": row["Address"],
            "Location": row["Location"],
            "Price": row["Price"],
            "Cuisine": row["Cuisine"],
            "Longitude": float(row["Longitude"]) if row["Longitude"] is not None else None,
            "Latitude": float(row["Latitude"]) if row["Latitude"] is not None else None,
            "PhoneNumber": str(row["PhoneNumber"]) if row["PhoneNumber"] is not None else None,
            "Url": row["Url"],
            "WebsiteUrl": row["WebsiteUrl"],
            "Award": row["Award"],
            "GreenStar": int(row["GreenStar"]) if row["GreenStar"] is not None else None,
            "FacilitiesAndServices": row["FacilitiesAndServices"],
            "Description": row["Description"],
        },
        "google_info": {
            "google_rating": row["google_rating"] if isinstance(row["google_rating"], (int, float)) else None,
            "google_reviews": row["google_reviews"] if isinstance(row["google_reviews"], (int, float)) else None,
        },
    }
    structured_data.append(structured_entry)

# Convert to JSON format
json_output = json.dumps(structured_data, indent=4, ensure_ascii=False)

# Save to file
json_file_path = "../data/structured_michelin_google_mongo_null.json"
with open(json_file_path, "w") as json_file:
    json_file.write(json_output)

# Provide the download link
json_file_path
```
